```python
import os
import random
import shutil
import pandas as pd

# Function to select and copy random audio files
def select_and_copy_files(source_folder, destination_folder, csv_file):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Create an empty DataFrame to store selected rows
    selected_df = pd.DataFrame(columns=df.columns)

    # Iterate through subfolders
    for folder in os.listdir(source_folder):
        folder_path = os.path.join(source_folder, folder)

        # Check if the path is a directory
        if os.path.isdir(folder_path):
            # Get a list of audio files in the subfolder
            audio_files = [file for file in os.listdir(folder_path) if file.endswith('.wav')]

            # Check if the subfolder has at least 10 audio files
            if len(audio_files) >= 10:
                # Select 10 random audio files
                selected_files = random.sample(audio_files, 10)

                # Copy the selected files to the destination folder
                for file in selected_files:
                    source_file_path = os.path.join(folder_path, file)
                    destination_file_path = os.path.join(destination_folder, file)
                    shutil.copyfile(source_file_path, destination_file_path)

                    # Find the corresponding row in the DataFrame and append it to a new DataFrame
                    selected_row = df[(df['filename'] == file[:-4]+".mp3") & (df['ebird_code'] == folder)]
                    selected_df = pd.concat([selected_df, selected_row], ignore_index=True)

    # Save the new DataFrame to a new CSV file
    selected_df.to_csv("C:/Users/ASHESH/Downloads/selected_data.csv", index=False)
    print(selected_df)

# Define source and destination folders, and the CSV file
source_folder = "C:/Users/ASHESH/Downloads/archive"
destination_folder = "C:/Users/ASHESH/Downloads/BirdDataset"
csv_file = "C:/Users/ASHESH/Downloads/archive/train_mod.csv"

# Create an empty DataFrame to store selected rows
#selected_df = pd.DataFrame(columns=['filename', 'ebird_code'])

# Call the function
select_and_copy_files(source_folder, destination_folder, csv_file)
```
